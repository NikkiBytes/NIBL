```python
for file in glob.glob("bbx*.log"):
    #print(file)
    sub= "sub-"+file.split("_")[1]
    session = "ses-"+file.split("_")[3]
    run = "run-"+file.split("_")[2].split("n")[1]
    
    if "post" in session or "pre" in session:
        print(session)
        try:
            df = pd.read_csv(file, engine="python", sep="\\t", header=None)
            df.columns = ["onset", "data", "keypress"]
            df = df[["onset", "data"]]
            start = df[df["data"].str.contains("start key press")]
            tasty = df[df["data"].str.contains("image=SL.jpg|image=CO.jpg")]
            ntasty = df[df["data"].str.contains("image=USL.jpg|image=UCO.jpg")]
            water = df[df["data"].str.contains("image=water.jpg")]
            WW = df[df["data"].str.contains("Level injecting via pump at address 0")]
            TT = df[df["data"].str.contains("Level injecting via pump at address 1")]
            NtNt = df[df["data"].str.contains("Level injecting via pump at address 2")]
            rinse = df[df["data"].str.contains("Level RINSE")]

            start = start[["onset"]]
            start_time = float(start.iat[0,0])
    
            tasty = tasty[["onset"]] - start_time
            tasty.onset = tasty.onset.round(4)
            tasty["duration"] = 6
            tasty["cue"] = 1
    
    
            ntasty = ntasty[["onset"]] - start_time
            ntasty.onset = ntasty.onset.round(4)
            ntasty["duration"] = 6
            ntasty["cue"] = 1
    
    
            water = water[["onset"]] - start_time
            water.onset = water.onset.round(4)
            water["duration"] = 6
            water["cue"] = 1
    
            WW = WW[["onset"]] - start_time
            WW.onset = WW.onset.round(4)
            WW["duration"] = 6
            WW["cue"] = 1
    
            TT = TT[["onset"]] - start_time
            TT.onset = TT.onset.round(4)
            TT["duration"] = 6
            TT["cue"] = 1
    
            NtNt = NtNt[["onset"]] - start_time
            NtNt.onset = NtNt.onset.round(4)
            NtNt["duration"] = 6
            NtNt["cue"] = 1
    
            rinse = rinse[["onset"]] - start_time
            rinse.onset = rinse.onset.round(4)
            rinse["duration"] = 3
            rinse["cue"] = 1
    
            outpath = "/Users/nikkibytes/Documents/BBx_logs/output/"
    
            files2make=['rinse', 'TT', 'NtNt', 'WW', 'Tcue', 'Ucue', 'Ncue']
        

            for name in files2make:
                filename = "%s_%s_%s_%s.txt"%(sub, session, name, run)
                if name == "rinse":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename )
                    else:
                        rinse.to_csv(outpath+filename, sep='\t', header=None, index=None)
                elif name == "TT":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename )
                    else:
                        TT.to_csv(outpath+filename, sep='\t', header=None, index=None)
                elif name == "NtNt":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename)
                    else:
                        NtNt.to_csv(outpath+filename, sep='\t', header=None, index=None)
                elif name == "WW":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename)
                    else:
                        WW.to_csv(outpath+filename, sep='\t', header=None, index=None)
                elif name == "Tcue":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename)
                    else:
                        tasty.to_csv(outpath+filename, sep='\t', header=None, index=None)
                elif name == "Ucue":
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename)
                    else:
                        ntasty.to_csv(outpath+filename, sep='\t', header=None, index=None)
                else:
                    if rinse.empty:
                        print("empty file, ", file, "\n", filename)
                    else:
                        water.to_csv(outpath+filename, sep='\t', header=None, index=None)
            

        except pd.errors.EmptyDataError:
            print("EMPTY FILE ", file)     
```
