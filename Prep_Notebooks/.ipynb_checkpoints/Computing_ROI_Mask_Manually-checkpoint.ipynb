```python
fmri_data = fmri_img.get_data()
print(fmri_data.shape)
```

Now we will select features using T-test: here we use the student's t-test(scipy.stats.ttest_ind)- an established method to determine whether 2 distributions have a different mean value. It can be used to compare voxel time-series from two different experimental conditions. If the time series distribution is similar in the 2 conditions, then the voxel is not very interesting to discriminate on.

```python
import numpy as np
from scipy import stats 
## will have to modify labels for specifc subject labels, right now I am taking one fo the files
## and resizing it to fit -- not valid using to test 
_, p_values = stats.ttest_ind(fmri_data[..., labels=="H2O"],
                             fmri_data[..., labels=="unapp"],
                             axis=-1)
fmri_data
```

```python
log_p_values = -np.log10(p_values) #use a log scale for p-values 

print(log_p_values)
# NAN values to zero 
log_p_values[np.isnan(log_p_values)] = 0. 
log_p_values[log_p_values > 10. ] = 10. 




```
