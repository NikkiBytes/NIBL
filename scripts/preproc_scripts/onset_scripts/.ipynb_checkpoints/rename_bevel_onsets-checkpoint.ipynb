```python
import glob
import pandas as pd
import os
```

```python
df = pd.read_csv('/Users/nikkibytes/Documents/Test/Bevel.txt', sep='\t', header=None)
df.columns = ["subID", "bevelID"]

```

```python
df.head()
```

```python
list = []
for value in df['bevelID']:
    bev_id = value.split('_')[0]
    list.append(bev_id)
df['orgID'] = list
df = df.drop(columns="bevelID")
df.head()
```

```python
df = df.set_index("orgID")
df
```

```python
df.index = df.index.str.strip()
df.index.values
```

```python
df.loc['bevel01']
```

```python
os.chdir('/Users/nikkibytes/Documents/Bevel_Onsets')
missing_subjects = [] 
for file in glob.glob('*txt'):
    bev = file.split("_")[0]
    #print(file)
    try:
        sub = df.loc[bev,"subID"].strip() 
        run = file.split("_")[2].split('0')[1].split('.')[0]
        task = file.split("_")[1]
        print("ORIGINAL FILE: ", file)
        newfilename = "%s_task-%s_run-%s.txt"%(sub, task, run)
        print("NEWFILE: ", newfilename)
        os.rename(file, newfilename)
    except KeyError:
        missing_subjects.append(bev)
```

```python
missing_subjects
```

```python
i = set(missing_subjects)
missing_subs=[]
for t in i:
    if 'sub' not in t:
        missing_subs.append(t)
```
